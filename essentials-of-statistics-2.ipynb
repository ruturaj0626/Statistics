**3. Descriptive Statistics Measures**
   - Explain what each measure is and when to use them.
   - Provide Python code examples for calculating mean, median, mode, variance, standard deviation, and range.

```python
import numpy as np

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

mean = np.mean(data)
median = np.median(data)
mode = np.argmax(np.bincount(data))
variance = np.var(data)
std_dev = np.std(data)
data_range = np.ptp(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Range: {data_range}")
```

**4. Data Visualization**
   - Explain the importance of data visualization.
   - Provide code examples for creating histograms, box plots, and scatter plots.

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()

# Scatter plot example using random data
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
